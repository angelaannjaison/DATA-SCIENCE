```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris=load_iris()
#taking two cols since we need only two parameters for drawing a scattered plot
x=iris.data[:,:2]#1st parameter:all rows,2nd parameter:2 cols(0th index,1st index)
y=iris.target
plt.scatter(x[:,0],x[:,1],c=y,cmap="prism")
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.show()
```

```python
km=KMeans(n_clusters=3)#3 clses in target,so 3 clusters
km.fit(x)
centers=km.cluster_centers_
print(centers)
```

```python
new_label=km.labels_
print(new_label)#predicted label
print(y)#actual label
```

```python
fig,axes=plt.subplots(1,2,figsize=(16,5))
#actual plot=box 0--------------predicted box=box 1
axes[0].scatter(x[:,0],x[:,1],c=y,cmap='prism')
axes[0].set_xlabel('sepal length')
axes[0].set_ylabel('sepal width')
axes[0].set_title('Actual')
axes[1].scatter(x[:,0],x[:,1],c=new_label,cmap='jet')
axes[1].set_xlabel('sepal length')
axes[1].set_ylabel('sepal width')
axes[1].set_title('Predicted')
plt.show()
```
